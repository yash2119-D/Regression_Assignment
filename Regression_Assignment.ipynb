# **Regression Assignment Answers**

## **1. What is Simple Linear Regression?**
Simple Linear Regression is a statistical method that models the relationship between a dependent variable and one independent variable using a straight-line equation. The goal is to determine how the independent variable influences the dependent variable.

---

## **2. What are the key assumptions of Simple Linear Regression?**
The key assumptions are:
1. **Linearity**: The relationship between the dependent and independent variable should be linear.
2. **Independence**: The observations should be independent of each other.
3. **Homoscedasticity**: The variance of residuals should be constant across all values of the independent variable.
4. **Normality**: The residuals should be normally distributed.
5. **No Multicollinearity**: Since this is simple regression, there is only one independent variable.

---

## **3. What does the coefficient \( m \) represent in the equation \( Y = mX + c \)?**
The coefficient \( m \) represents the slope of the regression line, which indicates the rate at which the dependent variable changes with respect to the independent variable.

---

## **4. What does the intercept \( c \) represent in the equation \( Y = mX + c \)?**
The intercept \( c \) represents the value of the dependent variable when the independent variable is zero.

---

## **5. How do we calculate the slope \( m \) in Simple Linear Regression?**
The slope is determined using the least squares method, which minimizes the differences between the actual and predicted values.

### **Python Code:**
```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 3, 5, 7, 11])

# Linear Regression Model
model = LinearRegression()
model.fit(X, Y)

# Get slope (m) and intercept (c)
m = model.coef_[0]
c = model.intercept_

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")

# Plotting
plt.scatter(X, Y, color='blue', label="Data Points")
plt.plot(X, model.predict(X), color='red', label="Regression Line")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

---

## **6. What is the purpose of the least squares method in Simple Linear Regression?**
The least squares method helps find the best-fitting regression line by minimizing the sum of squared differences between actual and predicted values.

---

## **7. How is the coefficient of determination (\( R^2 \)) interpreted in Simple Linear Regression?**
The \( R^2 \) value measures how well the independent variable explains the variance in the dependent variable. A higher \( R^2 \) value indicates a better fit.

### **Python Code:**
```python
r_squared = model.score(X, Y)
print(f"RÂ² Value: {r_squared}")
```

---

## **8. What is Multiple Linear Regression?**
Multiple Linear Regression is an extension of Simple Linear Regression that uses multiple independent variables to predict a dependent variable.

---

## **9. What is the main difference between Simple and Multiple Linear Regression?**
- **Simple Linear Regression**: One independent variable.
- **Multiple Linear Regression**: More than one independent variable.

---

## **10. What are the key assumptions of Multiple Linear Regression?**
1. **Linearity**: Each independent variable should have a linear relationship with the dependent variable.
2. **No Multicollinearity**: Independent variables should not be highly correlated with each other.
3. **Homoscedasticity**: The variance of residuals should remain constant.
4. **Normality**: Residuals should be normally distributed.

---

## **11. What is heteroscedasticity, and how does it affect Multiple Linear Regression?**
Heteroscedasticity occurs when the variance of residuals is not constant, leading to unreliable estimates.

### **Python Code to Detect Heteroscedasticity:**
```python
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

# Fit a regression model
X_sm = sm.add_constant(X)
model_sm = sm.OLS(Y, X_sm).fit()

# Perform Breusch-Pagan test
_, pval, _, _ = het_breuschpagan(model_sm.resid, X_sm)
print(f"Breusch-Pagan test p-value: {pval}")
```

## **16. How does the intercept in a regression model provide context for the relationship between variables?**
The intercept represents the expected value of the dependent variable when all independent variables are zero. In some cases, it provides meaningful insights, while in others, it may not be interpretable, especially if zero is outside the range of the data.

## **17. What are the limitations of using \( R^2 \) as a sole measure of model performance?**
- \( R^2 \) does not indicate whether the model is appropriate.
- It does not account for overfitting.
- It does not assess the correctness of model assumptions.
- A high \( R^2 \) does not mean a model will make good predictions.

---

## **18. How would you interpret a large standard error for a regression coefficient?**
A large standard error suggests that the coefficient estimate is uncertain and may not be statistically significant. This could indicate multicollinearity, small sample size, or high variability in the data.

---

## **19. How can heteroscedasticity be identified in residual plots, and why is it important to address it?**
Heteroscedasticity can be identified using residual plots, where residual variance increases or decreases systematically. Addressing it is important to improve model accuracy and avoid biased statistical inferences.

### **Python Code for Residual Plot:**
```python
import seaborn as sns
import statsmodels.api as sm

# Plot residuals
residuals = model_sm.resid
sns.residplot(x=model.predict(X), y=residuals, lowess=True, color="blue")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

This means that adding independent variables is not improving the model. Adjusted \( R^2 \) accounts for the number of predictors, so a large difference indicates that some variables are not contributing useful information.

## **23. How does polynomial regression differ from linear regression?**
- Linear regression fits a straight line.
- Polynomial regression fits a curved line by adding higher-order terms.

---

## **24. When is polynomial regression used?**
It is used when data exhibits a nonlinear relationship that cannot be captured effectively by a simple linear regression model.

---

## **25. What is the general equation for polynomial regression?**
The general equation is:
\[
Y = b_0 + b_1X + b_2X^2 + ... + b_nX^n
\]
where higher-degree terms allow for more flexible curve fitting.

---

## **26. Can polynomial regression be applied to multiple variables?**
Yes, polynomial regression can be extended to multiple variables by applying polynomial transformations to each independent variable.

---

## **27. What are the limitations of polynomial regression?**
- High-degree polynomials can lead to overfitting.
- Interpretation becomes complex.
- Requires careful selection of the degree to avoid unnecessary complexity.

---

## **28. What methods can be used to evaluate model fit when selecting the degree of a polynomial?**
- **Cross-validation**: Tests model performance on different data subsets.
- **Adjusted \( R^2 \)**: Evaluates if additional polynomial terms improve the model.
- **Mean Squared Error (MSE)**: Measures average squared differences between actual and predicted values.

````python
### **Python Code for 5th number Quetion:**
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 3, 5, 7, 11])

# Linear Regression Model
model = LinearRegression()
model.fit(X, Y)

# Get slope (m) and intercept (c)
m = model.coef_[0]
c = model.intercept_

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")

# Plotting
plt.scatter(X, Y, color='blue', label="Data Points")
plt.plot(X, model.predict(X), color='red', label="Regression Line")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
````

````python
### **Python Code for 7th number Question:**
r_squared = model.score(X, Y)
print(f"RÂ² Value: {r_squared}")
````

````python
#Python code for 11th number question......
### **Python Code to Detect Heteroscedasticity:**
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

# Fit a regression model
X_sm = sm.add_constant(X)
model_sm = sm.OLS(Y, X_sm).fit()

# Perform Breusch-Pagan test
_, pval, _, _ = het_breuschpagan(model_sm.resid, X_sm)
print(f"Breusch-Pagan test p-value: {pval}")
````

````python
#Python code for 14th number of Question.....
### **Python Code for Polynomial Regression:**
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Create Polynomial Regression Model
poly_model = make_pipeline(PolynomialFeatures(2), LinearRegression())
poly_model.fit(X, Y)

# Plot
plt.scatter(X, Y, color='blue')
plt.plot(X, poly_model.predict(X), color='red')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Polynomial Regression")
plt.show()
````
